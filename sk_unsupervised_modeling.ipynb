## Imports and config

```python
import sklearn.datasets
from sklearn.cluster import KMeans

mpl.style.use("ggplot_updated.mplstyle")
```

## Dataset loading

```python
X, y = sklearn.datasets.load_wine(return_X_y = True, as_frame = True)
```

## Modeling

### K-means
K-means is an classification/clustering algorithm that is based on the ecluedian distance between different points. K stands for the number of clusters that you want to divide your dataset in. The algorithm starts with centroids (centers of the different clusters), then classifies each point to belong to its closest centroid. Afterwards each centroid is shifted to the middle/mean of its cluster, with this new centroid again all points are classified to the closest centroid. This process repeats until the centroids only move in marginal way, or until the limit of repetitions is reached.

In 2D the representation is quite straightforward, however the dimension depends on how many features are in the dataset. Let's start with an easy 2D representation:

```python
X_test = X[["alcohol", "malic_acid"]]
model = KMeans(n_clusters = 3, init = "k-means++")
y_pred = model.fit_predict(X_test)
```

K-means++ basically means that your initial selection of centroid coordinates is less random, which means you can get to convergence faster. In this 2D representation we only base ourselves on 2 features (alcohol and malic_acid), the clusters seems to show a nicely categorized areas. The centroid points at the end of the kmeans are shown in red.
